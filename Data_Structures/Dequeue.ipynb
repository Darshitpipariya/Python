```python
class DeQueue:
    """ queue operations """
    def __init__(self):
        self.queue=[]
        print("Queue is created")
    def isEmpty(self):
        return self.queue==[]
    def size(slef):
        return len(self.queue)
    def enqueue_end(self,item):
        self.queue.append(item)
    def enqueue_front(self,item):
        self.queue.insert(0,item)
    def dqueue_end(self):
        print("element to be deque is",self.queue[-1])
        self.queue.pop()
    def dqueue_front(self):
        print("element to be deque is",self.queue[0])
        self.queue.pop(0)
```
